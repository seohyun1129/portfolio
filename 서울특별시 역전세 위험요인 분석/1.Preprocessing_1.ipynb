- 보증금이 3년치 월세를 초과하는 *보증부월세*의 경우 월별, (지역별, 면적별) 전월세전환율을 이용하여 전세로 환산
  - df.loc[(df['전월세구분']=='월세') & (df['계약기간'] >= 36.0) & (df['계약기간']*df['임대료(만원)'] > df['보증금(만원)'])]
  - 논문 기반 전처리에 존재하는 내용 -> *모두 월세로 분류됨* -> **생략!**
  - #df.loc[~((df['전월세구분'] == '월세') & (df['계약기간(개월)'] >= 36.0) & (df['계약기간(개월)'] * df['임대료(만원)'] <= df['보증금(만원)'])), '전월세구분'] = '월세'

- 전세 -> 월세 공식으로 바꿔보면
- 전세보증금 = 월세 * 계약기간(개월) / 전환율 + 월세보증금
- https://yellowsugar.tistory.com/69

```python
mask1 = ((df['전월세구분'] == '월세') & (df['계약기간(개월)'] >= 36.0) & (df['계약기간(개월)'] * df['임대료(만원)'] <= df['보증금(만원)']))
df.loc[mask1, '전월세구분'] = '전월세 전환'
df.loc[mask1, '보증금(만원)'] = df.loc[mask1, '임대료(만원)'] * df.loc[mask1, '계약기간(개월)'] / df.loc[mask1, '전월세전환율'] + df.loc[mask1, '보증금(만원)']
df.loc[mask1,'임대료(만원)'] = 0
```

```python
df['전월세구분'].value_counts()
```

- 보증금(만원)이 3,400만원 이하인 경우도 월세에 가깝다고 봐서 월세로 판단

```python
df.loc[df['보증금(만원)'] <= 3400 , '전월세구분'] = '월세'
```

##### 신축여부 (EDA:건물용도별 건축년도별 전세보증금)
- ~ 5년 : 신축
- 6 ~ 10년 : 준신축
- 11년 ~ : 구축

```python
df['건축년도'].isna().sum()
```

```python
# 신축여부 컬럼 생성
df['신축여부'] = " "
df.loc[(df['계약년도'] - df['건축년도']) <= 5, '신축여부'] = "신축"
df.loc[((df['계약년도'] - df['건축년도']) > 5) & ((df['계약년도'] - df['건축년도']) <= 10), '신축여부'] = "준신축"
df.loc[(df['계약년도'] - df['건축년도']) > 10, '신축여부'] = "구축"
```
